```python
'''
Code by Albert Au Yeung
Jun 3, 2018  
'''
def generate_ngrams(s, n):
    s = s.lower()
    s = re.sub(r'[^a-zA-Z0-9\s]', '', s) 
    tokens = [token for token in s.split(" ") if token != ""]
    ngrams = zip(*[tokens[i:] for i in range(n)])
    return [" ".join(ngram) for ngram in ngrams]
```

```python
bigram_words = {}
unigram_words = {}
trigram_words = {}
#uni_gram_bi_words = {}
backward_word = {}
forward_backward_word = {}
```

```python
with open('/content/gdrive/My Drive/TextMining/gutenberg-poetry-v001.ndjson', 'r') as f:
    for line in f:
        peotry = json.loads(line)
        sentence = peotry['s']
        sentence = sentence.lower()
        #sentence = re.sub(r'[^a-z\s]', '', sentence)
        #sentence = re.sub(r"\b[a-z]\b", '', sentence)
        #sen =sentence.split()
        '''
        Uni Gram Dictionary
        '''
        token_sentence = sentence.split()
        #print(token_sentence)
        for word in token_sentence:
            word = word.translate(str.maketrans('', '', string.punctuation))
            if word in unigram_words.keys():
                unigram_words[word] += 1
            else:
                unigram_words[word] = 1
        
        '''
        Bi Gram Dictionary
        '''
        bi_token_sentence = generate_ngrams(sentence, n=2)
        #print(bi_token_sentence)
        for word in bi_token_sentence:
            word = word.translate(str.maketrans('', '', string.punctuation))
            if word in bigram_words.keys():
                bigram_words[word] += 1
            else:
                bigram_words[word] = 1
            if word in forward_backward_word.keys():
                forward_backward_word[word] += 1
            else:
                forward_backward_word[word] = 1
        
        '''
        Tri Gram Dictionary
        '''
        tri_token_sentence = generate_ngrams(sentence, n=3)
        #print(tri_token_sentence)
        for word in tri_token_sentence:
            word = word.translate(str.maketrans('', '', string.punctuation))
            if word in trigram_words.keys():
                trigram_words[word] += 1
            else:
                trigram_words[word] = 1
            
        
        '''Backward Word Dictionary'''
        
        rev = ' '.join(token_sentence[-1::-1])
        bi_token_sentence = generate_ngrams(rev, n=2)
        #token_sentence = rev.split()
        #print(token_sentence)
        for word in bi_token_sentence:
            word = word.translate(str.maketrans('', '', string.punctuation))
            if word in backward_word.keys():
                backward_word[word] += 1
            else:
                backward_word[word] = 1
            if word in forward_backward_word.keys():
                forward_backward_word[word] += 1
            else:
                forward_backward_word[word] = 1
        
        
        peotries.append(peotry)
        
        #x = x + 1
        #if (x == 50000):
        #    break
```
